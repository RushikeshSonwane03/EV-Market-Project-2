## Pair Plot for Pairwise Relationships

```python
# Create a pairplot to visualize the pairwise relationships between different attributes
sns.pairplot(dataset, hue='Would you prefer replacing all your vehicles to Electronic vehicles?',  diag_kind='kde', corner=True)
plt.show()
```

# Co-Relation

```python
# Create a heatmap to visualize the correlation between attributes
plt.figure(figsize=(10,5))
sns.heatmap(dataset.corr(), annot=True)
plt.show()
```

## Age

```python
ax = sns.displot(dataset['Age'], kde = True)
plt.show()
```

```python
# Create a box plot to visualize the distribution of age for different marital status
sns.boxplot(y='Would you prefer replacing all your vehicles to Electronic vehicles?', x='Age', data=dataset)
plt.title('Distribution of Age by EV Preference')
plt.show()
```

```python
# Create a violin plot to compare Age by Preference
sns.violinplot(x="Age", y="Would you prefer replacing all your vehicles to Electronic vehicles?", data=dataset)
plt.show()
```

## Marital Status

```python
dataset['Marital Status'].describe()
```

```python
# Plotting the Preference status with respect to Marrital Status
sns.countplot(x ='Marital Status', hue = 'Would you prefer replacing all your vehicles to Electronic vehicles?',  data = dataset)
plt.show()
```

## Education

```python
# Plotting the Preference status with respect to Marrital Status
sns.countplot(x ='Education', hue = 'Would you prefer replacing all your vehicles to Electronic vehicles?',  data = dataset, palette = 'Set2')
plt.show()
```

## No. of Family Members

```python
ax = sns.displot(dataset['No. of Family members'], kde = True)
```

```python
# Plotting the Preference status with respect to Marrital Status
sns.countplot(x ='No. of Family members', hue = 'Would you prefer replacing all your vehicles to Electronic vehicles?',  data = dataset, palette = 'Set2')
plt.show()
```

```python
# Create scatter plot with hue
sns.violinplot(x='No. of Family members', y='Annual Income', hue='Would you prefer replacing all your vehicles to Electronic vehicles?', data=dataset)

# Add title and axis labels
plt.title('Family members and Annual Income, by Preferance')
plt.xlabel('No. of Family members')
plt.ylabel('Annual Income')
plt.show()
```

```python
# Create scatter plot with hue
sns.violinplot(y='No. of Family members', x='Preference for wheels in EV', hue='Would you prefer replacing all your vehicles to Electronic vehicles?', data=dataset)

# Add title and axis labels
plt.title('Family members and Preference for wheels')
plt.xlabel('Preference for wheels')
plt.ylabel('Family members')
plt.show()
```

## Profession

```python
# Create scatter plot with hue
sns.violinplot(x='Profession', y='Annual Income', hue='Would you prefer replacing all your vehicles to Electronic vehicles?', data=dataset)

# Add title and axis labels
plt.title('Family members and Annual Income, by Preferance')
plt.xlabel('No. of Family members')
plt.ylabel('Annual Income')
plt.show()
```

## Overall

```python
sns.countplot(x=dataset['Would you prefer replacing all your vehicles to Electronic vehicles?'])
plt.title("Count Plot of Preferences for EV")
plt.xlabel('EV Preferences')
plt.ylabel('Count')
plt.show()
```

```python
# Create a pie chart to visualize the preference for wheels in EV
plt.figure(figsize=(5,5))
dataset['Preference for wheels in EV'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Preference for Wheels in EV')
plt.show()
```
